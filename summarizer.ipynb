```python
!pip install transformers
!pip install gradio
from transformers import pipeline
from transformers import pipeline


# Initialize the HuggingFace summarization pipeline
summarizer = pipeline("summarization")
#summarized = summarizer(to_tokenize, min_length=75, max_length=300)


def predict(prompt):
    completion = summarizer(prompt, min_length=75, max_length=300)
    return completion

import gradio as gr

demo = gr.Interface(fn=predict, inputs="text", outputs="text",title="Faiz ul haque Zeya gradio text summarization page",
description="This is text summarization app that uses transformer library to complete and gradio to display")

demo.launch(share=True)
```
